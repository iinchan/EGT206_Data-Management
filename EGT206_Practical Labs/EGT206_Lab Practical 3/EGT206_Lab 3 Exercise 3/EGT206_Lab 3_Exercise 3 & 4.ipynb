### Step 2. Read from student-mat.csv and assign it to a variable called df.

```python
df = pd.read_csv("student-mat.csv")
df.head()
```

### Step 3. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
# create a function for conditional statements
def majority(x):
  if x > 17:
      return "True" # legal if age is > 17
  else:
      return "False" # illegal if age is < 17
```

```python
df["legal_drinker"] = df["age"].apply(majority) # create a column called "legal_drinker" and apply the function by using the age
df.head()
```

# Ex 4 - Plotting

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame it should look like below.

```python
data = {
    "first_name": ["Jason", "Molly", "Tina", "Jake", "Amy", "John"],
    "last_name": ["Miller", "Jacobson", "Ali", "Milner", "Cooze", "Cooper"],
    "female": [0, 1, 1, 0, 1, 0],
    "age": [42, 52, 36, 42, 42, 36],
    "preTestScore": [4, 24, 31, 2, 3, 50],
    "postTestScore": [25, 94, 57, 62, 70, 80]
}

new = pd.DataFrame(data)
new
```

### Step 3. Create a Scatterplot with "preTestScore" as the x-axis and "postTestScore" as the y-axis.

```python
new.plot(kind = "scatter", x = "preTestScore", y = "postTestScore")

plt.title("preTestScore x postTestScore")
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")

plt.show()
```

### Step 4. Create a Histogram using column "Age" to find out the age distribution of participants in dataset.

```python
new["age"].plot(kind = "hist") # plot histogram based on age column

plt.title("Age Distribution")
plt.xlabel('Age')
plt.ylabel('Number of people')

plt.show()
```
